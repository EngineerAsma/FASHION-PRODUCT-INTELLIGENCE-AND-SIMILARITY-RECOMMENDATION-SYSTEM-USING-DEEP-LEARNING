STEP 3: EXPLORATORY DATA ANALYSIS

```python
import  pandas as pd

df = pd.read_csv("data\\cleaned_data.csv")
```

```python
df.isnull().sum()
```

```python
# 1. product count by gender
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.countplot(x="gender", data=df, order =df['gender'].value_counts().index)
plt.show()

print(df["gender"].value_counts())

print("INSIGHTS: \n" \
"1. Men dominates other categories with a  variety of 20K+ products \n" \
"2. Women category stands second by having nearly 18K products \n" \
"3. While others  including kids and unisex  possess low to minimal variety of products")
```

```python
# 2. gender vs colour distribution 
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='base_colour', hue='gender', order=df['base_colour'].value_counts().head(10).index)
plt.show()

print("INSIGHTS:\n" \
"1. Black is the predominant  color among  men , women and unisex products\n" \
"2. White scores second dominant color for both  the categories.")
```

```python
# 3. gender vs article type distribution

plt.figure(figsize=(12,6))
sns.countplot(data=df, x="article_type", hue="gender",
              order=df["article_type"].value_counts().head(15).index)
plt.xticks(rotation=45)
plt.title("Top Article Types by Gender")
plt.show()

print("INSIGHTS: \n" \
"1. Under men categroy, TShirts, Shirts and Shoes fill the stock at the largest\n" \
"2. Under Women category, Kurtaas, Tops, Handbags and Heels products fill the stock at the largest")
```

```python
#4. distribution of master category
plt.figure(figsize=(10,6)) 
sns.countplot(x="master_category", data=df, order = df['master_category'].value_counts().index)
plt.show()

print("INSIGHTS:\n" \
"1. Apparel products such as garments, clothing are high compared to other master categories\n"
"2. The most  important master categories are Apparels, Accessories, Footwear and Personal care ")
```

```python
# 5. distribution of article type
top_article = df['article_type'].value_counts().head(15)
plt.figure(figsize=(10,6))
sns.barplot(x=top_article.values, y=top_article.index)
plt.show()

print("INISGHTS: \n" \
"1. The above chart shows the  top 15 article types among all other varieties of products\n" \
"2. This helps us identify the  area of investment and need for quick restocking of several products at the earliest")
```

```python
#6. color  distribution across  categories
top_colors = df["base_colour"].value_counts().head(15)
plt.figure(figsize=(10,6))
sns.barplot(x=top_colors.values, y=top_colors.index)
plt.show()

print("INSIGHTS:\n" \
"1. The above bar plot shows the top 15 colors of all other prodcuts\n" \
"2. This helps us identify people interests, likes and recomend similar products")
```

```python
# 7. seasonal product trends
plt.figure(figsize=(10,6))
sns.countplot(x='season', data=df, order = df['season'].value_counts().index)
plt.show()

print("INSIGHTS:\n" \
"1. The products for Summer season are high compared to other seasons\n" \
"2. While winter products and fall season products are  used moderately ")
```

```python
# 8. year analysis
plt.figure(figsize=(10,6))
sns.countplot(x='year', data=df, order = df['year'].value_counts().index)
plt.show()

print("INSIGHTS: \n" \
"1. The year 2012 should  have the highest selling record as there are more number of products being stocked compared to all other years")
```

```python
# 9. brands analysis
top_brands = df["brand_name"].value_counts().head(15)
plt.figure(figsize=(10,6))
sns.barplot(x= top_brands.values, y=top_brands.index)
plt.show()

print("INSIGHTS: The  above graph shows the  top 15  brands chosen by  customers")
```

```python
# 10. top brands with respect to gender
plt.figure(figsize=(14, 8))
top10_brands = df['brand_name'].value_counts().head(10).index
filtered = df[df['brand_name'].isin(top10_brands)]
pd.crosstab(filtered['gender'], filtered['brand_name']).plot(kind='bar', stacked=True)
plt.show()

print("INSIGHTS: Brands across genders\n" \
"1. Mens' category -> ADIDAS, PUMA and NIKE \n" \
"2. Womens' category -> CATWALK and UNITED COLORS OF BENETTON ")
```

```python
#  11. price distriibution 
plt.figure(figsize=(8,5))
sns.histplot(df["price"], bins=50, kde=True)
plt.title("Price Distribution")
plt.show()

print("INSIGHTS:\n" \
"1. The majority of products fall roughly between ₹500–₹3000, indicating a focus on affordable to mid-range fashion items.\n" \
"2. A small number of high-priced outliers (up to ~₹30,000) exist but are rare. ")
```

```python
# 12. brand vs price distribution
top10_brands = df['brand_name'].value_counts().head(10).index
filtered = df[df['brand_name'].isin(top10_brands)]
avg_price = filtered.groupby("brand_name")["price"].mean().sort_values()

plt.figure(figsize=(8,6))
sns.barplot(x=avg_price.values, y=avg_price.index)
plt.title("Average Price per Brand")
plt.show()

print("INSIGHTS:\n" \
"1. It is clear that Womens' high-selling brands such Catwalk, Fabindia and  United colors of Benetton are of mid - high range yet most of them are affordable\n" \
"2. Mens' popular  brands such as adidas, puma and nike are the costliest brands of all")
```

```python
# 13. rating dsitribution
sns.countplot(x='myntra_rating', data=df)
```

```python
# 14. correlation heatmap between price, discounted price and  myntra rating
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.heatmap(df[["price", "discounted_price", "myntra_rating"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
