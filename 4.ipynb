STEP 4: CATEGORY SELECTION & FILTERING DATA (MEN / WOMEN / KIDS)

```python
import  pandas as pd
df = pd.read_csv("data\\cleaned_data.csv")
```

```python
# select womens category 

df_women = df[df["gender"] == "Women"].copy()
print(df_women)
```

```python
# inspect the women data
print(df_women.shape) 
print(df_women.isnull().sum())
df_women.reset_index(drop=True, inplace=True)
print(df_women['image_url'].isna().sum())
print(df_women.columns)
print(df_women.info())
print(df_women.describe())
```

```python
# check the distribution of article types for classification
print(df_women['article_type'].unique())
print("\n")
print(df_women['article_type'].value_counts().head(30))
```

```python
# select the top 10 classes from the varieties of  article types
selected_classes = [
    'Kurtas',
    'Handbags',
    'Tops',
    'Heels',
    'Tshirts',
    'Watches',
    'Flats',
    'Bra',
    'Wallets',
    'Sarees'
]
```

```python
# create a new  df with fitered men data containing only the selected classes
df_women_filtered = df_women[df_women['article_type'].isin(selected_classes)].copy()

df_women_filtered['article_type'].value_counts()
```

```python
# inspect the  filtered men data
print(df_women_filtered.shape)
print(df_women_filtered.head())
df_women_filtered.reset_index(drop=True, inplace=True)
print(df_women_filtered.head())
```

```python
# save the filtered men data  containing only the  selected classes as a csv file
df_women_filtered.to_csv("data\\filtered_women_data.csv", index=False) 
```
