```python
import pandas as pd
df = pd.read_csv("C:\\Users\\User\\.vscode\\FASHION PRODUCT INTELLIGENCE & SIMILARITY RECOMMENDATION SYSTEM USING DEEP LEARNING\\data\\extracted_data.csv")
```

```python
# Handle missing values in attributes like season, year, usage
print(df.isnull().sum())
print(df["display_categories"].fillna("Unknown", inplace=True))
print(df["display_categories"].value_counts())
print(df.dropna(subset = ['image_url'], inplace=True))
print(df.isnull().sum())

print(df.shape)
```

```python
df.isnull().sum()
```

```python
# remove duplicates or invalid product records
print(df.duplicated().any())
print(df.drop_duplicates(subset="id"))
```

```python
df.isnull().sum()
```

```python
# Ensure each product has at least one valid image URL
df = df[df['image_url'].str.startswith("http", na=False)]
```

```python
# Standardize category labels (e.g., Men / Women / Kids)

print(df["gender"].value_counts())

df["gender"] = df["gender"].str.strip().str.title()

df["gender"] = df["gender"].replace(
    {
        'Men' : 'Men',
        'Women' : 'Women',
        'Boys' : 'Kids',
        'Girls' : 'Kids',
        'Unisex' : 'Unisex'
    }
)

print(df["gender"].value_counts())

```

```python
print(df.shape)
df.reset_index(drop=True, inplace=True)
print(df.shape)
print(df.head())
df.to_csv("C:\\Users\\User\\.vscode\\FASHION PRODUCT INTELLIGENCE & SIMILARITY RECOMMENDATION SYSTEM USING DEEP LEARNING\\data\\extracted_structured_cleaned_data.csv", index = False)
```
