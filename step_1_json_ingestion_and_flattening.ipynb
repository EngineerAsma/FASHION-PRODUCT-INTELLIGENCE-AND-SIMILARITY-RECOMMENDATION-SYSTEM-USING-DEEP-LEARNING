STEP 1 : JSON INGESTION & FLATTENING

```python

import pandas as pd
import json
import glob
import os


folder_path = r'C:/Users/User/Downloads/styles/styles'

json_files = glob.glob(os.path.join(folder_path, '*.json'))

data_list = []

for file in json_files:
    try:
        with open(file, "r", encoding='utf-8') as f:
            json_data = json.load(f)

        data = json_data.get("data", {})

        extracted_data = {
            "id" : data.get("id"),
            "product_display_name" : data.get("productDisplayName"),
            "brand_name" : data.get("brandName"),
            "price" : data.get("price"),
            "discounted_price" : data.get("discountedPrice"),
            "myntra_rating": data.get("myntraRating"),
            "age_group": data.get("ageGroup"),
            "gender": data.get("gender"),
            "master_category" : data.get("masterCategory", {}).get("typeName"),
            "sub_category": data.get("subCategory", {}).get("typeName"),
            "article_type": data.get("articleType", {}).get("typeName"),
            "base_colour": data.get("baseColour"),
            "usage": data.get("usage"),
            "season": data.get("season"),
            "year": data.get("year"),
            "display_categories": data.get("displayCategories"),
            "image_url" : data.get("styleImages", {}).get("default", {}).get("imageURL")

        }
        data_list.append(extracted_data)

    except Exception as e:
        print(f"{file} file not found")

        

print(f" loaded {len(data_list)} files")

df = pd.DataFrame(data_list)

print(df)


        
```
