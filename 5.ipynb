STEP 5: IMAGE DATASET PREPARATION FOR CNN MODELLING -> IMAGE CLASSIFICATION

```python
import pandas as pd
# open the already filtered womens' data
df_women_filtered = pd.read_csv("data\\filtered_women_data.csv")

# open the csv file  containing the website url for the corresponding image -> to download the  images
image_links  = pd.read_csv("data\\images.csv")

print(df_women_filtered.shape)
print(image_links.head())
print(image_links.columns)
```

```python
# create a new column "id" to map both imagelinks and df_women_filtered to download the selected women images
image_links['id'] = image_links['filename'].str.replace('.jpg', '').astype(int)
print(image_links.head())
print(image_links.columns)
print(image_links.shape)
```

```python
# map df_women_filtered across image_links to get the link for the filtered women's data -> mapping based on join operation based on the common column "id" in both df_women_filtered and image_links
df_images = df_women_filtered.merge(image_links, on = 'id', how='inner') 

print(df_women_filtered.columns)

print(df_images.columns) # new df by mapping womens filtered dta with image links so that we get the links and filenames of the filtered women's data
print(df_images.shape)
print(df_images.head())
```

```python
# downloading image url
import requests
import os 
os.makedirs("women_images", exist_ok=True)


for i in range(len(df_images)):

    try:
        url = df_images.iloc[i]['link']

        filename = df_images.iloc[i]['filename']

        filepath = os.path.join("women_images", filename)

        if os.path.exists(filepath):
            continue

        response =  requests.get(url, timeout=5)

        if response.status_code == 200:

            with open(filepath, "wb") as f:
                f.write(response.content)

    except: 
        continue
```

```python
# check how many images been downloaded properly
import os

valid_files = os.listdir("women_images")
print(len(valid_files))

df_images = df_images[df_images['filename'].isin(valid_files)].copy()

df_images.reset_index(drop=True, inplace=True)

print(df_images.shape)
                    
```

```python
print(df_images.head())
df_images.to_csv("data\\image_dataset_for_modelling.csv", index=False)
```
