STEP 7: IMAGE EMBEDDING EXTRACTION (FEATURE EXTRACTION)

```python
#1. load the saved model

import tensorflow as tf
from tensorflow.keras.models import load_model

model = load_model("efficientnet_model.keras")
```

```python
#2. create feature extractor model 
# model's last layer -> class probabilities
# models second last layer -> features
from tensorflow.keras.models import Model

feature_extractor = Model(inputs = model.input, outputs = model.layers[-2].output)

```

```python
# check the images folder
import os
len(os.listdir("women_images"))
```

```python
# extract  features / embeddings for all images ans store its corresponding paths to locate the image in folder
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.efficientnet import preprocess_input
import numpy as np
import os
import pickle

features = []
paths = []

dataset_path = r"D:\PROJECTS\FASHION PRODUCT\notebooks\women_images"

for file in os.listdir(dataset_path):
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):
        
        img_path = os.path.join(dataset_path, file)

        try:
            img = image.load_img(img_path, target_size=(224,224))
            img = image.img_to_array(img)

            img = np.expand_dims(img, axis=0)
            img = preprocess_input(img)  

            fe = feature_extractor.predict(img, verbose=0)

            features.append(fe.flatten())
            paths.append(img_path)

        except:
            print("Error:", img_path)

features = np.array(features)

print("Final shape:", features.shape)

pickle.dump(features, open("features_embeddings.pkl", "wb"))
pickle.dump(paths, open("paths.pkl", "wb"))
```
